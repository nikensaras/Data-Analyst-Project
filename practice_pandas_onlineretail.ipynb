## Jumlah Nominal dari Pembatalan Transaksi

```
sukses = df[df['Total_Price'] > 0]
gagal = df[df['Total_Price'] < 0]

nilai_sukses = sukses['Total_Price'].sum()
nilai_gagal = gagal['Total_Price'].sum()  # Nilainya akan negatif
absolute_gagal = abs(nilai_gagal)
net = nilai_sukses - absolute_gagal
persentase = (absolute_gagal / nilai_sukses) * 100
persentase
```
